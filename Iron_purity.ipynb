```python
def detect_outliers_iqr_features(data,feature):
    feature_data=data[feature]
    q1=np.percentile(feature_data,25)
    q3=np.percentile(feature_data,75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upperbound=q3+1.5*iqr
```

```python
feature_name=['% Iron Feed','% Silica Feed','Starch Flow','Amina Flow','Ore Pulp Flow','Ore Pulp Density','Flotation Column 04 Air Flow','Flotation Column 05 Air Flow','Flotation Column 06 Air Flow','Flotation Column 07 Air Flow','Flotation Column 01 Level','Flotation Column 02 Level','Flotation Column 03 Level','Flotation Column 04 Level','Flotation Column 05 Level','Flotation Column 06 Level','Flotation Column 07 Level','% Silica Concentrate']
for i in feature_name:
    outliers=detect_outliers_iqr_features(df,i)
    print(i,':',outliers)
```

```python
df = df.drop(['date','Ore Pulp pH','Flotation Column 01 Air Flow', 'Flotation Column 02 Air Flow', 'Flotation Column 03 Air Flow','% Iron Concentrate'], axis=1)
df.head()
```

```python
X = df.drop(['% Silica Concentrate'], axis=1)
Y = df['% Silica Concentrate']
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = pd.DataFrame(min_max_scaler.fit_transform(X), columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.30, random_state=42) 
X_train
```

```python

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
_ = reg.fit(X_train, Y_train)
predictions = reg.predict(X_test)
predictions
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(Y_test, predictions)
error
```

```python
plt.scatter(Y_test, predictions)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print('MAE is',mean_absolute_error(Y_test,predictions))
print('MSE is',mean_squared_error(Y_test,predictions))
```
